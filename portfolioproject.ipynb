```python
#Импортирт библиотек, необходимых для работы  
import pandas as pd 
import seaborn as sns 
import matplotlib 
import matplotlib.pyplot as plt 
import numpy as np
plt.style.use("ggplot")
from matplotlib.pyplot import figure
%matplotlib inline 
matplotlib.rcParams['figure.figsize']=(12,8)
```

```python
#Чтение данных
df=pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
#пропущенные значения в процентах
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#Тип данных в каждом столбце
df.dtypes
```

```python
df.head()
```

```python
#сортировка данных по столбцу gross
df=df.sort_values(by=['gross'], inplace=False,ascending=False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
df.drop_duplicates()
df.head()
```

```python
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('budget vs Gross earnings')
plt.xlabel('gross earnings')
plt.ylabel('budget for film')
plt.show()
```

```python
sns.regplot(x='budget',y='gross',data=df,scatter_kws={"color":"red"},line_kws={"color":'blue'})
```

```python
#корреляцилнная матрица между всеми числовыми столбцами 
df.corr(method='pearson')
```

```python
#высокая корреляция между budget и gross
correlation_matrix=df.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.title('correlation matrix for numeric features')
plt.xlabel(' movie features')
plt.ylabel('movie features')
plt.show()
```

```python

df.head()
```

```python
#каждому значению типа object присваиваем числовое значение для проверки корреляции по всем столбцам
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
df.head()
```

```python
correlation_matrix=df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.title('correlation matrix for numeric features')
plt.xlabel(' movie features')
plt.ylabel('movie features')
plt.show()
```

```python
#корреляционная матрица по всем столбцам
df_numerized.corr()
```

```python
correlation_mat=df_numerized.corr()

corr_pairs=correlation_mat.unstack()
corr_pairs
```

```python
#сортировка значении корреляционнной матрицы
sorted_pairs=corr_pairs.sort_values()
sorted_pairs
```

```python
#Отображение значении с высокой корреляцией (больше 0,5)
high_corr=sorted_pairs[(sorted_pairs)>0.5]
high_corr
```

```python
#Отображениие на 15 лучших компаний по доходу
CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```
